```python
import warnings
warnings.simplefilter(action='ignore', category=Warning)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('features.csv')
pd.set_option('mode.use_inf_as_na', True)
df.fillna(0, inplace=True)
#df['mime'] = LabelEncoder().fit_transform(df['mime'])
pln = df[(df['class'] == 0) & (df['algorithm'] == 'aes') & (df['filetype'] == 'txt')]
aes = df[df['class'] == 1 & (df['filetype']=='txt') & (df['algorithm']=='aes')]
bf  = df[df['class'] == 1 & (df['filetype']=='txt') & (df['algorithm']=='bf')]
cha = df[df['class'] == 1 & (df['filetype']=='txt') & (df['algorithm']=='cha')]
des3 = df[df['class'] == 1 & (df['filetype']=='txt') & (df['algorithm']=='des3')]
rc4 = df[df['class'] == 1 & (df['filetype']=='txt') & (df['algorithm']=='rc4')]
```

```python
def box(col):
    plt.boxplot([pln[col], aes[col], bf[col], cha[col], des3[col], rc4[col]], showfliers=False);
    plt.xticks([1,2,3,4,5,6], ['plaintext', 'aes', 'bf', 'cha', 'des3', 'rc4'])
    plt.xlabel('encryption')
    plt.ylabel(col)
    plt.title('%s for txt files' % col)
    plt.grid()
    plt.show()
```

```python
box('mean')
box('autocorr')
box('pi')
box('bigrams')
box('mse')
box('kurtosis')
box('entropy')
box('stdev')
box('chi')
box('p')
box('hmean')
box('skew')
```
