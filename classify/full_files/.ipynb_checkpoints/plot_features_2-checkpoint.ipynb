```python
import warnings
warnings.simplefilter(action='ignore', category=Warning)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('features.csv')
pd.set_option('mode.use_inf_as_na', True)
df.fillna(0, inplace=True)
df['mime'] = LabelEncoder().fit_transform(df['mime'])
pln = df[(df['class'] == 0) & (df['algorithm'] == 'aes')]
enc = df[df['class'] == 1]
```

```python
for m in np.unique(df['mime']):
  print(m)
```

```python
for col in df.columns:
    if col in ['filetype', 'filename', 'mime', 'algorithm', 'class']:
        continue   

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18,6))# create 2x2 array of subplots
    pln.boxplot(col, by='filetype',showfliers=False, ax=axes[0], rot=90)
    enc.boxplot(col, by='filetype',showfliers=False, ax=axes[1], rot=90)
    axes[0].set_title('plaintext', size=24)
    axes[0].set_ylabel(col, size=18)
    axes[0].set_xlabel('filetype', size=18)
    axes[1].set_title('encrypted', size=24)
    axes[1].set_ylabel('')
    axes[1].set_xlabel('filetype', size=18)

    plt.suptitle('')
    plt.tight_layout()
    
    yp = axes[0].get_yticks()
    ye = axes[1].get_yticks()
    
    if (yp[-1] - yp[0]) > (ye[-1] - ye[0]):
        axes[0].set_yticks(yp)
        axes[1].set_yticks(yp)
    else:
        axes[0].set_yticks(ye)
        axes[1].set_yticks(yp)
    
    axes[1].set_yticklabels([])
    
    plt.show()
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(18,18))# create 2x2 array of subplots

pln.boxplot('mean', by='filetype',showfliers=False, ax=axes[0][0], rot=90)
pln.boxplot('autocorr', by='filetype',showfliers=False, ax=axes[1][0], rot=90)
pln.boxplot('pi', by='filetype',showfliers=False, ax=axes[2][0], rot=90)
pln.boxplot('bigrams', by='filetype',showfliers=False, ax=axes[3][0], rot=90)

enc.boxplot('mean', by='filetype',showfliers=False, ax=axes[0][1], rot=90)
enc.boxplot('autocorr', by='filetype',showfliers=False, ax=axes[1][1], rot=90)
enc.boxplot('pi', by='filetype',showfliers=False, ax=axes[2][1], rot=90)
enc.boxplot('bigrams', by='filetype',showfliers=False, ax=axes[3][1], rot=90)

plt.suptitle('')
plt.tight_layout()   

for r in range(3):
    for c in range(2):
        axes[r][c].set_xlabel('')
        axes[r][c].set_xticklabels([])
        
for r in range(1,4):
    for c in range(2):
        axes[r][c].set_title('')

axes[0][0].set_title('plaintext', size=24)
axes[0][1].set_title('encrypted', size=24)
axes[3][0].set_xlabel('filetype', size=18)
axes[3][1].set_xlabel('filetype', size=18)

axes[0][0].set_ylabel('mean', size=18)
axes[1][0].set_ylabel('autocorr', size=18)
axes[2][0].set_ylabel('pi', size=18)
axes[3][0].set_ylabel('bigrams', size=18)

plt.tight_layout()   
for r in range(4):
    yp = axes[r][0].get_yticks()
    ye = axes[r][1].get_yticks()
    
    if (yp[-1] - yp[0]) > (ye[-1] - ye[0]):
        axes[r][0].set_yticks(yp)
        axes[r][1].set_yticks(yp)
    else:
        axes[r][0].set_yticks(ye)
        axes[r][1].set_yticks(yp)
    
    axes[r][1].set_yticklabels([])
    axes[r][0].tick_params(axis='y', labelsize=12)

plt.tight_layout(w_pad=2, h_pad=0)
plt.show()
```
