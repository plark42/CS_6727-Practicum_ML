```python
import warnings
warnings.simplefilter(action='ignore', category=Warning)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('features.csv')
pd.set_option('mode.use_inf_as_na', True)
df.fillna(0, inplace=True)

pln = df[(df['class'] == 0) & (df['algorithm'] == 'aes')]
enc = df[df['class'] == 1]
```

```python
np.unique(df['mime'])
```

```python
for col in df.columns:
    if col in ['filetype', 'filename', 'mime', 'algorithm', 'class']:
        continue   

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18,6))# create 2x2 array of subplots
    pln.boxplot(col, by='filetype',showfliers=False, ax=axes[0])
    enc.boxplot(col, by='filetype',showfliers=False, ax=axes[1])
    plt.suptitle('')
    axes[0].set_title('plaintext')
    axes[1].set_title('encrypted')
    axes[0].set_ylabel(col)
    axes[1].set_ylabel(col)
    plt.tight_layout()
    plt.show()
```
